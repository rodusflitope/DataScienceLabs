Queremos predecir si el consumidor X comprara el producto Y en la semana Z por lo que es conveniente agrupar los datos de esta forma, es decir, cada registro representa una semana con su data y el label es si el cliente compra el producto Y en esta semana siendo 1 si lo compra y 0 si no lo compra. Es importante notar que al pasar los registros a semanas la data del cliente y los productos no cambia, solo las transacciones se deben manejar para crear features que capturen la informaci√≥n que aportan como por ejemplo agregar la cantidad que se compro cierto objeto en la semana. Para esto revisamos el rango de fechas con el objetivo de saber como dividir

```python
fecha_inicio = transacciones['purchase_date'].min()
fecha_fin = transacciones['purchase_date'].max()

print(f"Rango de fechas en transacciones: {fecha_inicio} a {fecha_fin}")
```

Separamos el dataset por semana, ya que, el problema pide predecir las compras por semana por esto nos combiene tener el dataset de tal forma que producto-cliente-semana define un registro y que el label sea si en ese registro se compra o no el producto por ese usuario esa semana. Podemos notar que hay demasiadas combinaciones producto, semana, consumidor por lo que hacemos negative sampling para que sea manejable (hacemos sampleo de casos donde no se compr√≥ el objeto por ese consumidor en esa semana en vez de tomar todas las opciones posibles). Es importante notar que no tomamos el order_id en este paso pues representa muchas categor√≠as que pueden ser muy dificiles de manejar para el modelo.

```python
from sklearn.base import BaseEstimator, TransformerMixin

RATIO_NEG_POS = 2  # Ratio de casos negativos a positivos para hacer el negative sampling

class WeeklyDatasetBuilder(BaseEstimator, TransformerMixin):
    def __init__(self, ratio_neg_pos: int = 1, random_state: int | None = RANDOM_STATE):
        self.ratio_neg_pos = ratio_neg_pos
        self.random_state = random_state

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        transacciones = X['transacciones'].copy()
        clientes = X['clientes']
        productos = X['productos']

        transacciones['purchase_date'] = pd.to_datetime(transacciones['purchase_date'])
        fecha_inicio = transacciones['purchase_date'].min()
        transacciones['week'] = ((transacciones['purchase_date'] - fecha_inicio).dt.days // 7).astype(int)

        # Casos positivos que salen de las transacciones reales (por eso para todos el label es 1)
        positivos = (
            transacciones.groupby(['customer_id', 'product_id', 'week'])
            .size()
            .reset_index(name='count')
        )
        positivos['purchased'] = 1

        # Universo cliente-producto-semana
        semanas_totales = range(transacciones['week'].min(), transacciones['week'].max() + 1)
        clientes_activos = transacciones['customer_id'].unique()
        productos_vendidos = transacciones['product_id'].unique()
        universe = pd.MultiIndex.from_product(
            [clientes_activos, productos_vendidos, semanas_totales],
            names=['customer_id', 'product_id', 'week']
        ).to_frame(index=False)

        dataset_completo = universe.merge(positivos, on=['customer_id', 'product_id', 'week'], how='left')
        dataset_completo['purchased'] = dataset_completo['purchased'].fillna(0)

        # Negative sampling estratificado por cliente (para que sea realista sino podria sobrerepresentar clientes inactivos)
        casos_positivos = dataset_completo[dataset_completo['purchased'] == 1]
        casos_negativos = dataset_completo[dataset_completo['purchased'] == 0]

        negativos_sample = []
        for cliente in clientes_activos:
            neg_cliente = casos_negativos[casos_negativos['customer_id'] == cliente]
            pos_cliente = casos_positivos[casos_positivos['customer_id'] == cliente]
            # Calculo cuantos negativos tomar (si tengo muchos positivos y no quedan negativos, tomo todos los negativos)
            n_sample = min(len(pos_cliente) * self.ratio_neg_pos, len(neg_cliente))
            if n_sample > 0:
                neg_sample = neg_cliente.sample(n=int(n_sample), random_state=self.random_state)
                negativos_sample.append(neg_sample)

        # Caso borde porque en la pr√°ctica siempre hay negativos
        if len(negativos_sample) > 0:
            casos_negativos_sample = pd.concat(negativos_sample, ignore_index=True)
            dataset_completo = pd.concat([casos_positivos, casos_negativos_sample], ignore_index=True)
        else:
            dataset_completo = casos_positivos.copy()

        # A√±ado atributos de clientes y productos
        dataset_final = (
            dataset_completo
            .merge(clientes, on='customer_id', how='left')
            .merge(productos, on='product_id', how='left')
        )

        # Elimininar columna 'count' si existe porque no es necesaria, solo quiero saber si compr√≥ o no el producto
        if 'count' in dataset_final.columns:
            dataset_final = dataset_final.drop(columns=['count'])

        return dataset_final

# Re-construir dataset_final con el builder 
builder = WeeklyDatasetBuilder(ratio_neg_pos=RATIO_NEG_POS, random_state=RANDOM_STATE)
dataset_final_train = builder.fit_transform({
    'transacciones': train_trans,
    'clientes': clientes,
    'productos': productos,
})

dataset_final_val = builder.transform({
    'transacciones': val_trans,
    'clientes': clientes,
    'productos': productos,
})

dataset_final_test = builder.transform({
    'transacciones': test_trans,
    'clientes': clientes,
    'productos': productos,
})

print("Dataset final de entrenamiento info:")
print(dataset_final_train.info())
print("\nPrimeras filas:")
print(dataset_final_train.head())
print(f"\nDimensiones del dataset final de entrenamiento: {dataset_final_train.shape}")
```

Convertimos cada columna a algo procesable para el modelo. Notar que se deja fuera de esto a las columnas que definen cada registro como son week, product_id y client_id pues no agregan informaci√≥n. Para el caso de week lo usamos para agregar una feature de week c√≠clica, esto por 2 motivos, el primero es que week es s√≥lo un identificador de cada registro pero no captura estacionalidad en el a√±o lo que si aporta informaci√≥n, por esto se toma la semana respecto al a√±o (como estos datos son de un a√±o no afecta directamente pero la idea es que la pipeline generalice). El segundo punto a notar es que se uso seno y coseno, esto porque si se definieran de forma simple las semanas desde 1 a 52 para representar la semana del a√±o, el modelo entender√≠a que 1 y 52 estan muy lejos pero estan tan lejos como 1 de 2 por lo que se usa una funci√≥n c√≠clica para capturar esta cercan√≠a c√≠clica.

```python
import numpy as np
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler, FunctionTransformer
from sklearn.pipeline import Pipeline

cat_cols = [
    'region_id',
    'customer_type',
    'brand',
    'category',
    'sub_category',
    'segment',
    'package',
]
num_base_cols = [
    'size',
    'num_deliver_per_week',
    'num_visit_per_week',
]

# Funci√≥n que agrega codificaci√≥n c√≠clica de semana y remueve 'week' asi capturo estacionalidad pues cada registro ya separa por semana
def add_week_cyclical(X: pd.DataFrame) -> pd.DataFrame:
    Xc = X.copy()
    w = (Xc['week'] % 52).astype(float)
    Xc['week_sin'] = np.sin(2 * np.pi * w / 52.0)
    Xc['week_cos'] = np.cos(2 * np.pi * w / 52.0)
    return Xc.drop(columns=['week'])

# A√±adimos features
add_time = FunctionTransformer(add_week_cyclical)
num_cols = num_base_cols + ['week_sin', 'week_cos']

# Procesamos columnas categ√≥ricas y num√©ricas
preprocess = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), cat_cols),
        ('num', StandardScaler(), num_cols),
    ],
    remainder='drop',
    verbose_feature_names_out=False,
)

# Pipeline completo
prep_pipeline = Pipeline(steps=[
    ('add_time', add_time),
    ('preprocess', preprocess),
])

feature_cols = cat_cols + num_base_cols + ['week']

X_train = dataset_final_train[feature_cols]; y_train = dataset_final_train['purchased']
X_val   = dataset_final_val[feature_cols];   y_val   = dataset_final_val['purchased']
X_test  = dataset_final_test[feature_cols];  y_test  = dataset_final_test['purchased']

Xt_train = prep_pipeline.fit_transform(X_train)
Xt_val   = prep_pipeline.transform(X_val)
Xt_test  = prep_pipeline.transform(X_test)

print({'train_shape': Xt_train.shape, 'val_shape': Xt_val.shape, 'test_shape': Xt_test.shape})
```

Seleccionamos un `DecisionTreeClassifier` como modelo baseline. Esta elecci√≥n se debe a que es un modelo simple, r√°pido de entrenar e interpretable, lo que lo convierte en un excelente punto de partida para problemas de clasificaci√≥n. Al establecer `class_weight='balanced'`, ayudamos al modelo a manejar el desbalance de clases presente en los datos, esto para que no se vea sobreestimada la ventaja de los modelos que si manejan balance de clases con respecto a este baseline.

Aunque mostramos muchas m√©tricas, la m√©trica principal que usaremos es la PR AUC (Average Precision Score), elegida por su robustez en escenarios con clases desbalanceadas, complementada por el F1-Score para la clase positiva. Ambas m√©tricas se enfocan en la clase positiva y es por esto que fue interesante usarlas pues ambas se enfocan en la clase positiva, sobre todo la PR AUC pues adem√°s calcula sobre todos los umbrales siendo m√°s consistente que el f1 score y por eso se eleigi√≥ como m√©trica primaria

```python
from sklearn.metrics import classification_report, roc_auc_score, average_precision_score
import time

# Fruncion que automatiza el evauar modelos
def eval_model(name, pipe, X_tr, y_tr, X_va, y_va):
    start_time = time.time()
    pipe.fit(X_tr, y_tr)
    training_time = time.time() - start_time
    
    y_pred = pipe.predict(X_va)
    report = classification_report(y_va, y_pred, digits=4, output_dict=True)
    roc_auc = None
    pr_auc = None
    if hasattr(pipe, "predict_proba"):
        y_score = pipe.predict_proba(X_va)[:, 1]
        roc_auc = roc_auc_score(y_va, y_score)
        pr_auc = average_precision_score(y_va, y_score)
    print(f"\n{name} (Validaci√≥n):")
    print(classification_report(y_va, y_pred, digits=4))
    if roc_auc is not None:
        print(f"ROC AUC: {roc_auc:.3f} | PR AUC: {pr_auc:.3f}")
    print(f"Tiempo de entrenamiento: {training_time:.2f} segundos")
    return {
        'model': name,
        'f1_pos': report.get('1', report.get('1.0', {}))['f1-score'],
        'recall_pos': report.get('1', report.get('1.0', {}))['recall'],
        'precision_pos': report.get('1', report.get('1.0', {}))['precision'],
        'accuracy': report['accuracy'],
        'roc_auc': roc_auc,
        'pr_auc': pr_auc,
        'training_time_s': training_time,
    }
```

Es conocido por su velocidad y eficiencia en el uso de memoria, lo que lo hace ideal para datasets grandes y para una experimentaci√≥n m√°s r√°pida, como la optimizaci√≥n de hiperpar√°metros. Ideal para probar una opci√≥n m√°s moderna y r√°pida y como se comporta respecto a las otras.

```python
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier, ExtraTreesClassifier
from lightgbm import LGBMClassifier

from sklearn.base import TransformerMixin
class ToDense(TransformerMixin):
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X.toarray() if hasattr(X, 'toarray') else X

results = []
model_pipes = {}

# 1) Gradient Boosting (requiere densificar)
gb_pipe = Pipeline(steps=[
    ('prep', clone(prep_pipeline)),
    ('to_dense', ToDense()),
    ('clf', GradientBoostingClassifier(random_state=RANDOM_STATE)),
])
model_pipes["GradientBoostingClassifier"] = gb_pipe
results.append(eval_model("GradientBoostingClassifier", gb_pipe, X_train, y_train, X_val, y_val))

# 2) Random Forest
rf_pipe = Pipeline(steps=[
    ('prep', clone(prep_pipeline)),
    ('clf', RandomForestClassifier(random_state=RANDOM_STATE)),
])
model_pipes["RandomForestClassifier"] = rf_pipe
results.append(eval_model("RandomForestClassifier", rf_pipe, X_train, y_train, X_val, y_val))

# 3) Extra Trees
et_pipe = Pipeline(steps=[
    ('prep', clone(prep_pipeline)),
    ('clf', ExtraTreesClassifier(random_state=RANDOM_STATE)),
])
model_pipes["ExtraTreesClassifier"] = et_pipe
results.append(eval_model("ExtraTreesClassifier", et_pipe, X_train, y_train, X_val, y_val))

# Balance de clases para LightGBM
pos = int(y_train.sum())
neg = len(y_train) - pos
scale_pos_weight = neg / max(1, pos)

# 4) LightGBM
lgb_pipe = Pipeline(steps=[
    ('prep', clone(prep_pipeline)),
    ('clf', LGBMClassifier(
        random_state=RANDOM_STATE,
        scale_pos_weight=scale_pos_weight,
    )),
])
model_pipes["LGBMClassifier"] = lgb_pipe
results.append(eval_model("LGBMClassifier", lgb_pipe, X_train, y_train, X_val, y_val))

# Resumen y selecci√≥n del mejor modelo
df_res = pd.DataFrame(results)
df_res = df_res.sort_values(by=['pr_auc', 'f1_pos'], ascending=False)
print("\nResumen modelos (validaci√≥n):")
print(df_res)
```

Podemos ver que todos los modelos obtuvieron mejores m√©tricas que el modelo baseline y que el mejor este estos, en todas las m√©tricas, es el modelo `LGBMClassifier` teniendo mejor f1_score y PR_AUC que todos los otros modelos adem√°s de ser r√°pido para entrenarse, por esto es el modelo que seleccionamos para seguir pues es el mejor en resultados y es un muy buen candidato a optimizaci√≥n de hiperpar√°metros por grilla.

A partir de su an√°lisis anterior, se debe proceder a optimizar el rendimiento del modelo seleccionado mediante la optimizaci√≥n de sus hiperpar√°metros. Para ello, se espera que implementen `Optuna` para optimizar no solo los hiperpar√°metros del modelo, sino tambi√©n los de los preprocesadores utilizados (por ejemplo, OneHot Encoding, Scalers, etc.).

En nuestro caso, tras la comparaci√≥n de modelos, simplificamos la b√∫squeda para optimizar √∫nicamente **LightGBM** con `Optuna`, usando PR AUC en validaci√≥n como m√©trica objetivo.

Al desarrollar este proceso, deber√°n responder las siguientes preguntas clave como m√≠nimo:

- ¬øQu√© m√©trica decidieron optimizar y por qu√©?

- ¬øQu√© hiperpar√°metro tuvo un mayor impacto en el rendimiento de su modelo?

- ¬øCu√°nto mejor√≥ el rendimiento del modelo despu√©s de la optimizaci√≥n de hiperpar√°metros?

Por lo antes dicho, se decidi√≥ optimizar sobre PR AUC pues se enfoca en la clase positiva la que para este problema es mucho m√°s importante, es robusta a desbalance que se genera al hacer un buen negative sampling (lo que la puso sobre la ROC AUC) y aparte de esto calcula sobre todos los umbrales siendo robusta.

```python
import optuna
from sklearn.pipeline import Pipeline
from sklearn.base import clone
from sklearn.metrics import average_precision_score
from lightgbm import LGBMClassifier

def build_lgbm_pipe(trial):
    params = {
        'n_estimators': trial.suggest_int('n_estimators', 300, 1200),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.2, log=True),
        'num_leaves': trial.suggest_int('num_leaves', 31, 255),
        'subsample': trial.suggest_float('subsample', 0.4, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.4, 1.0),
        'reg_lambda': trial.suggest_float('reg_lambda', 0.0, 5.0),
        'random_state': RANDOM_STATE,
        'n_jobs': -1,
        'scale_pos_weight': scale_pos_weight,
    }
    clf = LGBMClassifier(**params)
    pipe = Pipeline(steps=[
        ('prep', clone(prep_pipeline)),
        ('clf', clf),
    ])
    return pipe

X_tr, y_tr = X_train, y_train
X_va, y_va = X_val, y_val


def objective(trial):
    pipe = build_lgbm_pipe(trial)
    pipe.fit(X_tr, y_tr)
    y_score = pipe.predict_proba(X_va)[:, 1]
    pr_auc = average_precision_score(y_va, y_score)
    trial.report(pr_auc, step=0)
    return pr_auc

sampler = optuna.samplers.TPESampler(seed=RANDOM_STATE)
study = optuna.create_study(direction='maximize', sampler=sampler)
study.optimize(objective, n_trials=100)

print("Mejor PR AUC (validaci√≥n):", study.best_value)
print("Mejores hiperpar√°metros:", study.best_params)
```

Entrenemos entonces el modelo con los mejores par√°metros para conseguir todas las m√©tricas

```python
best_params = study.best_params

# Crear pipeline con mejores hiperpar√°metros - solo usar par√°metros que existen en best_params
best_lgb = LGBMClassifier(
    objective='binary',
    random_state=RANDOM_STATE,
    n_jobs=-1,
    scale_pos_weight=scale_pos_weight,
    n_estimators=best_params['n_estimators'],
    learning_rate=best_params['learning_rate'],
    num_leaves=best_params['num_leaves'],
    subsample=best_params['subsample'],
    colsample_bytree=best_params['colsample_bytree'],
    reg_lambda=best_params['reg_lambda']
)

optuna_pipe = Pipeline(steps=[
    ('prep', clone(prep_pipeline)),
    ('clf', best_lgb)
])

# Entrenamiento y predicci√≥n
eval_model("LGBMClassifier Optimizado", optuna_pipe, X_train, y_train, X_val, y_val)
```

Con la optimizaci√≥n de par√°metros mejora muy ligera del orden de 0.002 en PR AUC pero debido a la velocidad del modelo, tiene sentido obtener estos par√°metros

```python
import shap

# Usaremos el mejor pipeline con LGBM optimizado
best_pipe = optuna_pipe

# Extraer el modelo y el preprocesador del pipeline
preprocessor = best_pipe.named_steps['prep']
model = best_pipe.named_steps['clf']

# Transformar los datos de validaci√≥n para obtener los nombres de las caracter√≠sticas
# Usamos una muestra para que SHAP no tarde tanto
X_val_sample = X_val.sample(10000, random_state=RANDOM_STATE)
y_val_sample = y_val.loc[X_val_sample.index]

X_val_transformed = preprocessor.transform(X_val_sample)
feature_names = preprocessor.named_steps['preprocess'].get_feature_names_out()

# Convertir a DataFrame para que SHAP muestre nombres de columnas
X_val_transformed_df = pd.DataFrame(X_val_transformed.toarray(), columns=feature_names)

# Crear el explainer de SHAP y calcular los valores
explainer = shap.TreeExplainer(model, feature_perturbation="interventional")
shap_values = explainer.shap_values(X_val_transformed_df, check_additivity=False)

# Visualizaci√≥n de la importancia global de las caracter√≠sticas
print("Importancia global de las caracter√≠sticas (SHAP):")
shap.summary_plot(shap_values, X_val_transformed_df, plot_type="bar", show=False)
plt.title("Importancia de Atributos (SHAP Value)")
plt.show()

print("\nDistribuci√≥n de la influencia de cada caracter√≠stica:")
shap.summary_plot(shap_values, X_val_transformed_df, show=False)
plt.show()

# An√°lisis de predicciones individuales
print("\nAn√°lisis de predicciones individuales:")

# Obtenemos los shap_values como un objeto que contiene toda la info necesaria
shap_values_obj = explainer(X_val_transformed_df)

# Obtenemos las probabilidades para encontrar los √≠ndices de inter√©s
y_probs = model.predict_proba(X_val_transformed_df)[:, 1]

# caso donde el modelo predijo una alta probabilidad de compra
high_prob_idx = np.argmax(model.predict_proba(X_val_transformed_df)[:, 1])
print(f"\n--- Ejemplo 1: Predicci√≥n con alta probabilidad de compra (√çndice: {high_prob_idx}) ---")
# La funci√≥n waterfall_plot ahora toma el objeto SHAP de la instancia que queremos explicar
shap.waterfall_plot(shap_values_obj[high_prob_idx], show=False)
plt.title(f"Explicaci√≥n para la predicci√≥n del registro {high_prob_idx}")
plt.show()

# caso donde el modelo predijo una baja probabilidad de compra
low_prob_idx = np.argmin(model.predict_proba(X_val_transformed_df)[:, 1])
print(f"\n--- Ejemplo 2: Predicci√≥n con baja probabilidad de compra (√çndice: {low_prob_idx}) ---")
shap.waterfall_plot(shap_values_obj[low_prob_idx], show=False)
plt.title(f"Explicaci√≥n para la predicci√≥n del registro {low_prob_idx}")
plt.show()

#Un caso de predicci√≥n incierta (probabilidad cercana a 0.5)
mid_prob_idx = np.argmin(np.abs(y_probs - 0.5))
print(f"\n--- Ejemplo 3: Predicci√≥n incierta o de confianza media (√çndice: {mid_prob_idx}) ---")
shap.waterfall_plot(shap_values_obj[mid_prob_idx], show=False)
plt.title(f"Explicaci√≥n para la predicci√≥n del registro {mid_prob_idx}")
plt.show()

# An√°lisis de interacciones
print("\nAn√°lisis de interacciones entre caracter√≠sticas:")
shap.dependence_plot("size", shap_values, X_val_transformed_df, interaction_index="sub_category_GASEOSAS", show=False)
plt.show()

print("\n--- Interacci√≥n entre 'week_cos' y 'sub_category_GASEOSAS' ---")
shap.dependence_plot("week_cos", shap_values, X_val_transformed_df, interaction_index="sub_category_GASEOSAS", show=False)
plt.show()
```

El modelo se puede explicar para predicciones espec√≠ficas utilizando los gr√°ficos de cascada de SHAP que muestran c√≥mo cada caracter√≠stica contribuye a mover la predicci√≥n desde un valor base hasta el valor final.

*   **Ejemplo 1: Predicci√≥n con alta probabilidad de compra:** En el primer gr√°fico de cascada, vemos una predicci√≥n que es significativamente m√°s alta que la base. Los factores que m√°s contribuyen a **aumentar** la probabilidad de compra (flechas rojas) son un valor espec√≠fico de `brand` y `size` (podr√≠a ser por la existencia de productos de ciertas marcas muy famosos por ejemplo las bebidas de la marca Cocacola Company).
